```python
sido_dic={'강원도':'강원','경기도':'경기','경상남도':'경남','경상북도':'경북','광주광역시':'광주','대구광역시':'대구',
          '대전광역시':'대전','서울특별시':'서울','전라남도':'전남','전라북도':'전북','제주특별자치도':'제주',
          '충청남도':'충남','충청북도':'충북','부산광역시':'부산','인천광역시':'인천','울산광역시':'울산','세종특별자치시':'세종'}
new_sido=[]
for sido in test_sido:
  new_sido.append(sido_dic[sido])
# new_sido
```

```python
t1=pd.DataFrame({'시도':new_sido})
# t1
```
