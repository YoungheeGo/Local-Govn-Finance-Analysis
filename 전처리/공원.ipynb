```python
sido_list=list(fin['시도'].unique())
print(f'시도개수: {len(sido_list)}')

sgg_dic=dict()
for sd in sido_list:
  sgg_dic[sd]=list(fin[fin['시도']==sd]['시군구'].unique())

for sd in sido_list:
  print(f'{sd}:{len(sgg_dic[sd])}')
```
