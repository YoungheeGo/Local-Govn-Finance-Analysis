```python
import numpy as np
# from dealib import dea.RTS, dea.Orientation,
from dea import RTS, Orientation, dea
import pandas as pd
```

```python
x_=pd.read_csv('/content/drive/MyDrive/.../dea_input.csv')
x_.iloc[:,2:]
```

```python
y_=pd.read_csv('/content/drive/MyDrive/.../dea_output.csv')
y_.iloc[:,2:]
```

```python
# Data from Benchmarking 4.6.1 Numerical example in R
x = np.array(x_.iloc[:,2:])
y = np.array(y_.iloc[:,2:])

result = dea(x, y, rts=RTS.vrs, orientation=Orientation.input) # vrs=0, input=0

print(result.eff)
print(result.slack)
print(result.lambdas)
```
